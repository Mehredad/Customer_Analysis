```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
# Load the dataset "Looker Ecommerce BigQuery Dataset"
events=pd.read_csv("events.csv")
inventory_items=pd.read_csv("inventory_items.csv")
order_items=pd.read_csv("order_items.csv")
orders=pd.read_csv("orders.csv")
products=pd.read_csv("products.csv")
users=pd.read_csv("users.csv")
```

```python
# Inspect the data
print("events\n", events.dtypes)
print("inventory_items\n", inventory_items.dtypes)
print("order_items\n", order_items.dtypes)
print("products\n", products.dtypes)
print("users\n", users.dtypes)
print(events.head())
print(inventory_items.head())
print(order_items.head())
print(products.head())
print(users.head())
```

```python
# Strip spaces from column names
events.columns = events.columns.str.strip()
inventory_items.columns = inventory_items.columns.str.strip()
order_items.columns = order_items.columns.str.strip()
orders.columns = orders.columns.str.strip()
products.columns = products.columns.str.strip()
users.columns = users.columns.str.strip()
```

```python
inventory_items.head(100)
```

```python
# Drop unnecessary columns only if they exist
columns_to_drop_events = ['sequence_number', 'session_id', 'created_at', 'ip_address', 'postal_code', 'browser', 'uri']
columns_to_drop_inventory_items = ['created_at', 'sold_at', 'product_sku', 'product_distribution_center_id']
columns_to_drop_order_items = ['inventory_item_id', 'status', 'created_at', 'shipped_at', 'delivered_at', 'returned_at']
columns_to_drop_orders = ['status', 'created_at', 'returned_at', 'shipped_at', 'delivered_at']
columns_to_drop_products = ['sku', 'distribution_center_id']
columns_to_drop_users = ['email', 'state', 'street_address', 'postal_code', 'latitude', 'longitude', 'created_at']

# Dropping columns for events DataFrame
for col in columns_to_drop_events:
    if col in events.columns:
        events.drop(col, axis=1, inplace=True)

# Dropping columns for inventory_items DataFrame
for col in columns_to_drop_inventory_items:
    if col in inventory_items.columns:
        inventory_items.drop(col, axis=1, inplace=True)

# Dropping columns for order_items DataFrame
for col in columns_to_drop_order_items:
    if col in order_items.columns:
        order_items.drop(col, axis=1, inplace=True)

# Dropping columns for orders DataFrame
for col in columns_to_drop_orders:
    if col in orders.columns:
        orders.drop(col, axis=1, inplace=True)

# Dropping columns for products DataFrame
for col in columns_to_drop_products:
    if col in products.columns:
        products.drop(col, axis=1, inplace=True)

# Dropping columns for users DataFrame
for col in columns_to_drop_users:
    if col in users.columns:
        users.drop(col, axis=1, inplace=True)
```

```python
# Summary statistics 
print(events.describe())
print(inventory_items.describe())
print(order_items.describe())
print(orders.describe())
print(products.describe())
print(users.describe())
```

```python
# Check the missing values 
print(events.isnull().sum())
print(inventory_items.isnull().sum())
print(order_items.isnull().sum())
print(products.isnull().sum())
print(users.isnull().sum())
```

```python
# Handle missing values
events['user_id'].fillna(-1, inplace=True)
events['city'].fillna('Unknown', inplace=True)
inventory_items['product_name'].fillna('Unknown', inplace=True)
inventory_items['product_brand'].fillna('Unknown', inplace=True)
products['name'].fillna('Unknow', inplace=True)
products['brand'].fillna('Unknow', inplace=True)
users['city'].fillna('Unknown', inplace=True)
```

```python
events['user_id'] = events['user_id'].astype('int32')
products['name'] = products['name'].astype('category')
```

# General Analysis of User Behaviors

```python
# Limit the geography to the United Kingdom
user_demographics_uk = users[users['country']=='United Kingdom']
user_demographics_uk.head()
```

```python
# Analyse users demographics 
user_demographics_uk.describe(include='all')
```

```python
# Event analysis (to understand the user interactions on the platform)
event_types = events['event_type'].value_counts()
print(event_types)
```

```python
# Product preferences
product_popularity = products.merge(inventory_items, left_on='id', right_on='product_id')
product_popularity = product_popularity.groupby('product_name').agg({'id_x': 'count'}).rename(columns={'id_x': 'count'}).sort_values(by='count', ascending=False)
print(product_popularity.head())
```

```python
traffic_sources_users = users['traffic_source'].value_counts()
traffic_sources_users_uk = user_demographics_uk['traffic_source'].value_counts()
traffic_sources_events = events['traffic_source'].value_counts()
print("Total traffic sources are: \n", traffic_sources_users, "\n")
print("Total traffic sources in the uk are: \n", traffic_sources_users_uk, "\n")
print("Total traffic sources events are: \n", traffic_sources_events, "\n")
```

# Tailoring Analysis for Toyify

```python
# Identify potential customers for Toyify
potential_customers = users[users['traffic_source'].str.contains('art|design|creative', case=False)]
print(potential_customers.head())
```

```python
toyify_related_products = products[products['name'].str.contains('toy|3D', case=False)]
related_product_performance = toyify_related_products.merge(
    inventory_items, left_on='id', right_on='product_id'
).groupby('product_name').agg({
    'id_x': 'count'
}).rename(columns={'id_x': 'count'}).sort_values(by='count', ascending=False)
print(related_product_performance.head())
```
